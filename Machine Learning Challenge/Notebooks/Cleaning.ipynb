```python
import numpy as np
import pandas as pd
from pandas.api.types import is_numeric_dtype
import sweetviz as sv
from scipy import stats
pd.set_option('display.max_rows', 100)
```

```python
#Reads Data and does some basic cleaning
df = pd.read_csv('./ClientBehaviourTraining.txt', sep='|')
df = df.drop(['Instalment_Amt_All', 'Current_Balance_All'], axis = 1)
df = df.dropna()
for index in range(9,81):
    if is_numeric_dtype(df.iloc[index]):
        q = df.iloc[index].quantile(0.99)
        df = df[df.iloc[index] < q]
df.head()
```

```python
#Generates Report
my_report = sv.analyze(df)
my_report.show_html()
```

```python
#Generates correlation coefficients to specific behaviour
behavname = 'VirtualCard' #Enter behaviour name here

behav = behavname + 'Take-up'
nobehav = behavname + 'NoTake-up'
appdf = df[df['BehaviourType'] == behav]
appdf = pd.concat([appdf, df[df['BehaviourType'] == nobehav]])
    
one_hot = pd.get_dummies(appdf['BehaviourType'])
one_hot = one_hot.drop(nobehav, axis = 1)
appdf = appdf.drop('BehaviourType', axis = 1)
appdf = appdf.join(one_hot)
appdf.corr().loc[behav].sort_values()
```

```python
app_report = sv.analyze(appdf)
app_report.show_html()
```

```python
feature_config = sweetviz.FeatureConfig(force_cat = ["Has_Deed"])
app_report = sv.compare_intra(appdf, appdf["AppTake-up"] == 1, ["Yes", "No"], None, feature_config)
app_report.show_html()
```
